# Tutorial 04: Full Digital Twin

End-to-end simulation loop.

```python
from deepiri_fuselk.sim import DigitalTwin

twin = DigitalTwin(grid_size=16)
state = twin.reset()
print(f'Initial: variance={state.heat_variance:.3f}, elm_p={state.elm_probability:.3f}')

for i in range(20):
    state = twin.step()
print(f'Final:   variance={state.heat_variance:.3f}, elm_p={state.elm_probability:.3f}')
print(f'Watchdog triggered: {state.watchdog_triggered}')
```
