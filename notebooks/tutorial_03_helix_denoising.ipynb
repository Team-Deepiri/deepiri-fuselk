# Tutorial 03: HELIX Focal Diagnostics

HQRM + focal heat map from synthetic ECE data.

```python
from deepiri_fuselk.sim import generate_ece_shot
from deepiri_fuselk.helix import run_hqrm, PhaseLockedTracker
from deepiri_fuselk.focal import focal_heatmap, from_hqrm
import matplotlib.pyplot as plt

shot = generate_ece_shot(32, seed=42)
hqrm = run_hqrm(shot.heat_field)
tracker = PhaseLockedTracker()
focal = from_hqrm(hqrm)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].imshow(shot.heat_field, cmap='hot'); ax[0].set_title('Raw')
ax[1].imshow(focal, cmap='viridis'); ax[1].set_title(f'Focal (var={hqrm.heat_variance:.3f})')
plt.tight_layout(); plt.show()
```
