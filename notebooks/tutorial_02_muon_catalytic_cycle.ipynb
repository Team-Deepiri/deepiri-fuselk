# Tutorial 02: Muon Catalytic Cycle

Run the muon rate network with photon/proton stripping.

```python
from deepiri_fuselk.muon import run_rate_network, RateNetworkParams, BREAKEVEN_FUSIONS

baseline = run_rate_network(params=RateNetworkParams(R_photon=0, R_proton=0))
boosted = run_rate_network(params=RateNetworkParams(R_photon=0.5, R_proton=0.3))
print(f'Baseline fusions/muon: {baseline.fusions_per_muon:.1f}')
print(f'Boosted fusions/muon:  {boosted.fusions_per_muon:.1f}')
print(f'Breakeven threshold:   {BREAKEVEN_FUSIONS}')
```
