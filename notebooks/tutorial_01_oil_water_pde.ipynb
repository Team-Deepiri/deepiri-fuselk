# Tutorial 01: Oil-Water PDE System

Solve the coupled plasma/vapor barrier equations.

```python
from deepiri_fuselk.physics import solve_oil_water_steady, peclet_criterion, PDEParameters
import matplotlib.pyplot as plt

result = solve_oil_water_steady(n_grid=64, max_iter=300)
state = result.state
print(f'Converged: {result.converged}, residual: {result.residual:.2e}')
print(f'Peclet number: {peclet_criterion(PDEParameters()):.2f}')

fig, ax = plt.subplots(1, 3, figsize=(12, 3))
ax[0].plot(state.x, state.n_p, label='plasma')
ax[0].plot(state.x, state.n_v, label='vapor')
ax[0].legend(); ax[0].set_title('Density profiles')
ax[1].plot(state.x, state.n_T); ax[1].set_title('Tritium')
ax[2].plot(state.x, state.T_p, label='T_p')
ax[2].plot(state.x, state.T_v, label='T_v')
ax[2].legend(); ax[2].set_title('Temperature')
plt.tight_layout(); plt.show()
```
